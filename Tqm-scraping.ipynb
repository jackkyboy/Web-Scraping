```python
import asyncio
from playwright.async_api import async_playwright, TimeoutError as PlaywrightTimeoutError

async def fetch_data():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)
        page = await browser.new_page()

        try:
            await page.goto("https://www.tqm.co.th/car-insurance/all", timeout=120000)
        except PlaywrightTimeoutError:
            print("Page load timeout exceeded")
            await browser.close()
            return

        try:
            await page.wait_for_load_state('networkidle', timeout=90000)
        except PlaywrightTimeoutError:
            print("Network idle timeout exceeded")
            await browser.close()
            return

        retry_count = 5
        for attempt in range(retry_count):
            try:
                await page.wait_for_selector('div[class*="rounded-2xl"]', timeout=30000)
                break
            except PlaywrightTimeoutError:
                print(f"Attempt {attempt + 1} of {retry_count} failed: Waiting for selector timeout exceeded")
                if attempt == retry_count - 1:
                    await browser.close()
                    return

        insurance_items = await page.query_selector_all('div[class*="rounded-2xl"]')

        for idx, item in enumerate(insurance_items):
            html = await item.inner_html()
            print(f"Insurance Item {idx+1} HTML:\n{html}\n")

            plan_name = await item.query_selector('h3[class*="truncate"]')
            price = await item.query_selector('div[class*="absolute"]')
            details = await item.query_selector('p[class*="text-body3"]')

            if plan_name and price and details:
                plan_name_text = await plan_name.inner_text()
                price_text = await price.inner_text()
                details_text = await details.inner_text()

                print(f"Item {idx+1}: Plan: {plan_name_text}, Price: {price_text}, Details: {details_text}")
            else:
                print(f"Item {idx+1}: Missing data, plan_name: {plan_name}, price: {price}, details: {details}")
        
        await browser.close()

asyncio.run(fetch_data())
```
